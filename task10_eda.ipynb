```python
df["Price"].hist(bins =20)
plt.title("Sale Price Distibution")
plt.show()
```

```python
plt.boxplot(df["Price"].dropna())
plt.title("Sale Price Boxplot")
plt.show()
```

```python
q1 = df["Price"].quantile(0.25)
q3 = df["Price"].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers = df[(df["Price"] < lower_bound) | (df["Price"] > upper_bound)]

outliers.shape
```

```python
df["outlier_flag"] = np.where((df["Price"] < lower_bound) | (df["Price"] > upper_bound), 1, 0)
```

```python
df["Price_capped"] = df["Price"]

df.loc[df["Price"] > upper_bound, "Price_cappped"] = upper_bound
df.loc[df["Price"] < lower_bound, "Price_capped"] = lower_bound
```

```python
corr = df.select_dtypes(include = np.number).corr()

plt.figure(figsize = (10,6))
sns.heatmap(corr, cmap = "coolwarm", annot = False)
plt.title("Correlation Matrix")
plt.show()
```
